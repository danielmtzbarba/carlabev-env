```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
from CarlaBEV.envs.utils import load_map

rgbmap, _, _ = load_map(size=128)
print(f"Image Size: {rgbmap.shape}")
plt.imshow(rgbmap)
plt.show()
```

```python
def draw_scene(points=[]):
    plt.imshow(rgbmap)
    for (x, y) in points:
        plt.scatter(x, y, s=5)
    plt.show()
```

```python
agent_0 = (850, 1080)
agent_f = (850, 900)
scene = [agent_0, agent_f]
draw_scene(scene)
```
